To view data from specific <b>columns</b> or <b>rows</b>, there are a few operations we can perform. First, lets look at accessing columns of data.

```python
df.columns # get column headers
```

```python
df['AGE'] # access a particular column
```

By accessing a particular column, we are returned a pandas <b>series</b>. A series is a one-dimensional ndarray with axis labels.

```python
type(df['AGE']) # notice this is of type series
```

By including a list of labels, we are specifying multiple columns to include

```python
df[['AGE', 'DEATH']] # access particular column(s)
```

Doing so returns a data frame object.

```python
type(df[['AGE', 'DEATH']]) # notice this is of type dataframe
```

```python
df[['AGE', 'SYSBP']] # get columns (sort of similar to dictionary reference to keys)
```

```python
df.iloc[[0,1]] # get rows
```

```python
df.iloc[[0,1], [0,1]] # get columns and rows (arg1 = columns and arg2 = rows)
```

```python
df.loc[[0,1], ['AGE', 'SYSBP']] # iloc is by index whereas loc is by label
```

```python
df.loc[0:5, ['AGE', 'SYSBP']] # slicing is possible too! although note that 5 is INCLUSIVE unlike regular slicing in python
```

```python
df.loc[0:5, 'AGE':'DIABETES'] # its also possible to slice labels!
```

```python
df.loc[df['AGE']==46] # only where age is 46
```

<h2>Indexing</h2>

Maybe I feel like integers aren't the best identifiers for my rows. Pandas allows us to change that!

```python
df.loc[0:5, 'AGE':'DIABETES'].set_index('AGE')
```

In this case, age is a pretty poor index as its not necessarily unique (note 46 appearing twice). Luckily, we did not modify the actual dataframe. To do so would require us to pass in an argument inplace=True such that our method now looks like <b>.set_index('AGE', inplace=True)</b>. You can also specify this when reading in the file: <b>df = pd.read_csv("framingham.csv", index_col='AGE')</b>

<h2>Modifying DataFrame</h2>

```python
df['SUM'] = df['SYSBP'] + df['DIABP'] # add a column to sum two others
cols = list(df.columns.values)
df = df[cols[0:2] + [cols[-1]] + cols[2:9]] # move sum
df.head(5)
```

```python
df = df.drop(columns=['SUM'])
df.head(5)
```

<h2>Filtering Data</h2>

We can filter data by columns. Note the ampersand and | and NOT 'and' or 'or'. We can not use and here as we are implicitly asking Python to convert (df['AGE']==48) and (df['DEATH']==1) to boolean values.

NumPy arrays (of length greater than 1) and Pandas objects such as Series do not have a boolean value

```python
# save new datframe
new_df = df.loc[(df['AGE']<37) & (df['DEATH']==1)] # df.loc[(df['AGE']==48) | (df['DEATH']==1)]
new_df.reset_index()
```

when we make a new dataframe, it can be helpful to reset the index

```python
new_df = new_df.reset_index()
new_df
```

The <b>re</b> library is very powerful for filtering

```python
import re
imdb = pd.read_csv("imdb.csv")
imdb
```

Here we wish to find where net or for are in the IMDB movie title. We specify that it is not case sensitive with flag I

```python
imdb.loc[(imdb['Title'].str.contains('NET|FoR', flags=re.I,regex=1))] 
```

<h2>Conditional Changes</h2>

Here we can specify that if some value is true, change said value to something else

```python
df.loc[df['AGE']<40, 'DEATH'] = 1 # if they are under 40, they die
# df.loc[df['AGE']<40, ['DEATH', ANYCHD]] = [1, 0]  HERE WE CAN CHANGE MULTIPLE VALUES
df
```

<h2>Aggregate Statistics (Groupby)</h2>

Here we can group by a specific category and gather statistics.

```python
means = df.groupby(['AGE']).mean()
means
```

Notice we have mean values for the different categories grouped by age

```python
means.sort_values('SYSBP', ascending=0) # sort by sysbp
```

```python
counts = df.groupby(['AGE']).count()
counts
```
